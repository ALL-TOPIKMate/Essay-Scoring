```python
from sentence_transformers import SentenceTransformer
import pandas as pd
import numpy as np
from sentence_transformers import util
model = SentenceTransformer('snunlp/KR-SBERT-V40K-klueNLI-augSTS')
```

```python
import pandas as pd
sentences = [
    "작년 대비 소비자물가가 33 %증가했다.",
    "작년 대비 소비자물가가 15% 증가했다.",
    "작년 대비 소비자물가가 33% 감소했다."
  ]
df = pd.DataFrame({"sentence" : sentences})
df['embeddings'] = df['sentence'].apply(lambda x: model.encode(x))
```

```python
df
```

```python
query = "작년 대비 소비자물가가 33% 증가했다."
query_encode = model.encode(query)

# 코사인 유사도 계산
cos_scores = util.pytorch_cos_sim(query_encode, df['embeddings'])[0]


df['cosine_similarity'] = cos_scores

# 결과 출력
#print(df[['sentence', 'cosine_similarity']])
df
```
